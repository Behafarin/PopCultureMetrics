```python
# Spotify app credentials
# Use your own credentials
client_id = ""
client_secret = ""

# Encode the client ID and secret for authorization header
auth_string = f"{client_id}:{client_secret}"
auth_bytes = auth_string.encode('utf-8')
auth_base64 = base64.b64encode(auth_bytes).decode('utf-8')

# Requesting an access token
def get_token():
    url = "https://accounts.spotify.com/api/token"
    headers = {
        "Authorization": f"Basic {auth_base64}",
    }
    data = {
        "grant_type": "client_credentials"
    }
    result = requests.post(url, headers=headers, data=data)
    json_result = result.json()
    return json_result['access_token']

# Function to search for an artist
def search_artist(token, artist_name):
    url = f"https://api.spotify.com/v1/search?q={artist_name}&type=artist"
    headers = {
        "Authorization": f"Bearer {token}"
    }
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Failed to fetch data for {artist_name}. Status code: {response.status_code}")
        return None

# Get the access token
token = get_token()

# Collect data for all artists
artist_data_list = []

for artist_name in artists[5600:]:
    print(f"Fetching data for {artist_name}...")
    artist_data = search_artist(token, artist_name)
    if artist_data and 'artists' in artist_data and artist_data['artists']['items']:
        artist = artist_data['artists']['items'][0]
        artist_dict = {
            'artist': artist.get('name'),
            'popularity': artist.get('popularity'),
            'followers': artist['followers']['total'],
            'genres': ", ".join(artist.get('genres', []))
        }
        artist_data_list.append(artist_dict)
    else:
        print(f"No data found for {artist_name}.")
    # Add a short delay to avoid rate limiting
    time.sleep(1)

# Save the data to a DataFrame
df = pd.DataFrame(artist_data_list)

# Save the data to a CSV file
df.to_csv("/content/drive/My Drive//spotify_artists_data.csv", index=False)
print("Data collection complete. Saved to spotify_artist_data.csv.")
```
