```python
import pandas as pd
from ucimlrepo import fetch_ucirepo
import requests
import base64
# Your API key (replace with your actual API key)
api_key = ""

# Function to search for news articles in content
def search_news_content(api_key, query, from_date=None, to_date=None, language='en', page_size=10):
    url = "https://newsapi.org/v2/everything"
    params = {
        'apiKey': api_key,
        'q': query,            # Search keyword in content
        'language': language,   # Specify language (English)
        'pageSize': page_size,  # Limit the number of articles returned
        'sortBy': 'relevancy'   # Sort results by relevance
    }

    if from_date:
        params['from'] = from_date
    if to_date:
        params['to'] = to_date

    response = requests.get(url, params=params)
    return response.json()
# Our sample to start with
artists = [
    "Taylor Swift", "The Weeknd", "Bad Bunny", "Drake", "Zach Bryan",
    "Sabrina Carpenter", "Billie Eilish", "Benson Boone", "Karol G", "Myke Towers",
    "Kendrick Lamar", "Ariana Grande", "Olivia Rodrigo", "Dua Lipa", "SZA",
    "Lana Del Rey", "Chappell Roan", "Lady Gaga", "Miley Cyrus", "Ed Sheeran",
    "Harry Styles", "Post Malone", "Beyoncé", "Rihanna", "Bruno Mars",
    "Justin Bieber", "Kanye West", "Eminem", "Lil Nas X", "Doja Cat",
    "Travis Scott", "J Balvin", "Shakira", "Rosalía", "Feid",
    "Peso Pluma", "Jung Kook", "Eslabon Armado", "The Kid LAROI", "Glass Animals",
    "Tones and I", "Lil Wayne", "Jay-Z", "Nicki Minaj", "BTS",
    "One Direction", "Katy Perry", "Demi Lovato", "Adele", "Coldplay"
]
news_results = []

for query in artists:
    news_data = search_news_content(api_key, query, from_date="2024-11-10", page_size=10)
    if news_data and 'articles' in news_data:
        for article in news_data['articles']:
            news_results.append({
                'artist': query,
                'title': article.get('title'),
                'published_at': article.get('publishedAt'),
                'content': article.get('content'),
                'url': article.get('url')
            })

# Convert to a DataFrame
df_news = pd.DataFrame(news_results)
df_news.to_csv("news_articles.csv", index=False)
```
